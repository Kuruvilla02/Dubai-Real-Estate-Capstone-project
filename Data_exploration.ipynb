### 1. Distribution of different transaction types over the entire dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 8))
sns.set(style="whitegrid")
sns.countplot(x='Transaction Type', data=df, palette='viridis')
plt.title('Distribution of Transaction Types', fontsize=16)
plt.xlabel('Transaction Type', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

### 2. Types of properties involved in transactions and their frequency.

```python
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.countplot(x='Property Type', data=df, hue='Transaction Type', palette='Set2', dodge=True)
plt.title('Transaction Types by Property Type', fontsize=16)
plt.xlabel('Property Type', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Transaction Type')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Transaction Size (sq.m)', y='Amount', data=df, hue='Transaction Type', palette='coolwarm')
plt.title('Transaction Size vs. Transaction Amount', fontsize=16)
plt.xlabel('Transaction Size (sq.m)', fontsize=14)
plt.ylabel('Transaction Amount', fontsize=14)
plt.legend(title='Transaction Type')
plt.show()
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
sns.lineplot(x='Transaction Date', y='Amount', hue='Transaction Type', data=df)
plt.title('Temporal Trends in Transactions', fontsize=16)
plt.xlabel('Transaction Date', fontsize=14)
plt.ylabel('Transaction Amount', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Transaction Type')
plt.show()
```

```python
import calendar

df['Transaction Month'] = df['Transaction Date'].dt.month_name()

heatmap_data = df.groupby(['Transaction Month', 'Property Type']).size().unstack(fill_value=0)

plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt='d', cbar_kws={'label': 'Frequency'})
plt.title('Heatmap of Transaction Frequencies by Month and Property Type', fontsize=16)
plt.xlabel('Property Type', fontsize=14)
plt.ylabel('Transaction Month', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(x='Amount', data=df, kde=True, bins=30, color='skyblue')
plt.title('Distribution of Property Prices', fontsize=16)
plt.xlabel('Property Price', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='Transaction Date', y='Amount', hue='Property Type', data=df)
plt.title('Price Trend by Property Type', fontsize=16)
plt.xlabel('Transaction Date', fontsize=14)
plt.ylabel('Property Price', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Property Type')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Property Type', y='Amount', data=df, palette='Set2')
plt.title('Boxplot of Property Prices by Type', fontsize=16)
plt.xlabel('Property Type', fontsize=14)
plt.ylabel('Property Price', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Property Size (sq.m)', y='Amount', data=df, hue='Property Type', palette='viridis')
plt.title('Price-Size Relationship', fontsize=16)
plt.xlabel('Property Size (sq.m)', fontsize=14)
plt.ylabel('Property Price', fontsize=14)
plt.legend(title='Property Type')
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.scatterplot(x='Longitude_Project', y='Latitude_Project', hue='Nearest Landmark',
                data=df, palette='Set2', s=100, alpha=0.7)
plt.title('Landmark Clusters', fontsize=16)
plt.xlabel('Longitude', fontsize=14)
plt.ylabel('Latitude', fontsize=14)
plt.legend(title='Nearest Landmark', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Room(s)', data=df, palette='pastel')
plt.title('Distribution of Room Counts', fontsize=16)
plt.xlabel('Number of Rooms', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Master Project', data=df, palette='Set1')
plt.title('Master Project Analysis', fontsize=16)
plt.xlabel('Master Project', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='Transaction Date', y='Amount', hue='Property Type', style='Registration type', data=df, markers=True, dashes=False)
plt.title('Segment-wise Price Trends', fontsize=16)
plt.xlabel('Transaction Date', fontsize=14)
plt.ylabel('Property Price', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Property Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
